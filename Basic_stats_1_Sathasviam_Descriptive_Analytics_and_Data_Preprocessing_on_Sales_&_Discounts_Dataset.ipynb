```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print("\nNumerical Columns:", numerical_columns)
```

```python
# Calculate descriptive statistics
descriptive_stats = df[numerical_columns].agg(['mean', 'median', 'std']).T
```

```python
# Calculate mode separately
descriptive_stats['mode'] = df[numerical_columns].mode().iloc[0] #	Extracts the first mode if there are multiple
```

```python
# Display results
print("\nDescriptive Statistics:")
print(descriptive_stats)
```

```python
# Interpretation of key findings
print("\nInterpretation:")
# Loops through each numerical column
for col in numerical_columns:
    print(f"\n {col}:")
    print(f"   - Mean: {descriptive_stats.loc[col, 'mean']:.2f}") #Formats numbers to 2 decimal places.
    print(f"   - Median: {descriptive_stats.loc[col, 'median']:.2f}")
    print(f"   - Mode: {descriptive_stats.loc[col, 'mode']:.2f}")
    print(f"   - Std Dev: {descriptive_stats.loc[col, 'std']:.2f}")

    if descriptive_stats.loc[col, 'mean'] > descriptive_stats.loc[col, 'median']:
        print("   - The data is right-skewed (has high-value outliers).")
    elif descriptive_stats.loc[col, 'mean'] < descriptive_stats.loc[col, 'median']:
        print("   - The data is left-skewed (has lower-value outliers).")
    else:
        print("   - The data is symmetrically distributed.")
```

# 2. Data Visualization

### 1. Histograms (Numerical Data)
#### Histograms help visualize data distribution.
#### The kde=True argument overlays a smooth density curve.
#### Analysis:
#### Symmetric distribution: Bell-shaped curve.
#### Right-skewed: Data has extreme high values.
#### Left-skewed: Data has extreme low values.
#### 2. Boxplots (Outlier Detection)
#### Boxplots help detect outliers and data spread.
#### Analysis:
#### Outliers are dots outside the whiskers.
#### IQR (Interquartile Range) shows data concentration.
#### Long whiskers indicate high variability.
### 3. Bar Charts (Categorical Data)
#### Bar charts help analyze category frequency.
#### Analysis:
#### Tallest bars represent dominant categories.
#### Evenly distributed bars suggest balanced data.
#### Rare categories may need special handling.

```python
# 	Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset from a CSV file
df = pd.read_csv("sales_data_with_discounts.csv")
# Display basic dataset information
print("\nDataset Overview:\n")
df.info()
```

```python
# Identify numerical and categorical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

print("\nNumerical Columns:", numerical_columns)
print("\nCategorical Columns:", categorical_columns)
```

# 2.1. HISTOGRAMS - To visualize the distribution of numerical variables

```python
# Set figure size
plt.figure(figsize=(15, 10))
```

```python
#Plot histograms for all numerical columns
#Loops through numerical_columns, providing both an index (i) and the column name (col)
#Creates a subplot grid for multiple boxplots - plt.subplot(rows, cols, index)
#Sets the title of the plot dynamically using f-string formatting (f"Boxplot of {col}"), where {col} is replaced with the current column name
for i, col in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns)//2 + 1, 2, i+1)
    sns.histplot(df[col], bins=20, kde=True, color='blue')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
print("\n **Histogram Analysis:**")
print("   - Check for skewness (left/right skewed).")
print("   - Identify outliers by observing tail distributions.")
```

# 2.2. BOXPLOTS - To detect outliers and check data spread

```python
# Set figure size
plt.figure(figsize=(15, 8))

# Create boxplots for numerical columns
for i, col in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns)//2 + 1, 2, i+1)
    sns.boxplot(y=df[col], color='orange')
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)

plt.tight_layout()
plt.show()

print("\n **Boxplot Analysis:**")
print("   - Outliers appear as individual points outside the whiskers.")
print("   - The interquartile range (IQR) shows data spread.")
```

# 2.3. BAR CHARTS - To analyze categorical column distribution

```python
# Set figure size
plt.figure(figsize=(15, 8))

# Plot bar charts for categorical columns
for i, col in enumerate(categorical_columns):
    plt.subplot(len(categorical_columns)//2 + 1, 2, i+1)
    sns.countplot(x=df[col], hue=df[col], palette="Set2", order=df[col].value_counts().index, legend=False)
    plt.xticks(rotation=45)
    plt.title(f"Bar Chart of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")

plt.tight_layout()
plt.show()

print("\n **Bar Chart Analysis:**")
print("   - Shows frequency of different categories.")
print("   - Helps in understanding dominant categories.")
```

# 3. Standardization of Numerical Variables
###  1. Identifying Numerical Columns
####  Selects only integer (int64) and float (float64) columns for scaling.
###  2. Applying Standardization
####  Uses StandardScaler() from sklearn to transform numerical columns.
####  Each value is transformed using the Z-score formula.
###  3. Comparing Distributions (Before & After)
####  Displays histograms for each numerical column before and after standardization.
####  Before Standardization: Features may have different scales.
####  After Standardization: All features are centered around 0 with unit variance.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
## Load dataset from a CSV file
df = pd.read_csv("sales_data_with_discounts.csv")

# Identify numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print("\nNumerical Columns:", numerical_columns)

# Display basic statistics BEFORE standardization
print("\n BEFORE Standardization:\n", df[numerical_columns].describe())
```

## Standardization (Z-Score Normalization)

```python
scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Display basic statistics AFTER standardization
print("\n AFTER Standardization:\n", df_standardized[numerical_columns].describe())
```

### Comparing Before and After Standardization with Histograms

```python
# Set figure size
plt.figure(figsize=(15, 10))

# Plot histograms BEFORE standardization
for i, col in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns), 2, 2*i+1)
    sns.histplot(df[col], bins=20, kde=True, color='skyblue')
    plt.title(f"Before Standardization: {col}")

    # Plot histograms AFTER standardization
    plt.subplot(len(numerical_columns), 2, 2*i+2)
    sns.histplot(df_standardized[col], bins=20, kde=True, color='orange')
    plt.title(f"After Standardization: {col}")

plt.tight_layout()
plt.show()

print("\n **Observations:**")
print("   - Before standardization, numerical features have different scales.")
print("   - After standardization, all features are centered around 0 with unit variance (mean = 0, std = 1).")
```

# 4. Conversion of Categorical Data into Dummy Variables

### 1. Identifying Categorical Columns
#### Selects columns with data type = object.
### 2. Applying One-Hot Encoding
#### Uses pd.get_dummies() to convert categories into binary columns (0 or 1).
### 3. Displaying Results
#### Shows original dataset and encoded dataset.
#### Lists newly created columns.

```python
# Import necessary libraries
import pandas as pd

# Upload dataset
df = pd.read_csv("sales_data_with_discounts.csv")

# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
print("\nCategorical Columns:", categorical_columns)

# Display first few rows BEFORE encoding
print("\n Dataset BEFORE One-Hot Encoding:\n")
print(df.head())
```

```python
# Apply One-Hot Encoding (Dummy Variables)
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)  # drop_first=True avoids dummy variable trap

# Display first few rows AFTER encoding
print("\n Dataset AFTER One-Hot Encoding:\n")
print(df_encoded.head())

# Show new column names
print("\n New Column Names after Encoding:\n", df_encoded.columns)
```
